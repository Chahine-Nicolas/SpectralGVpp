# East

```python
import numpy as np 
import json

path = "/lustre/fswork/projects/rech/dki/ujo91el/code/DSI_LIDAR_HD/dictio_results/results_dsi3d_east_global.json"
#path = "/lustre/fswork/projects/rech/dki/ujo91el/code/DSI_LIDAR_HD/dictio_results/results_dsi3d_moe_east_global.json"
with open(path, "r") as f:
    results_dict = json.load(f)
print(len(results_dict))

hit_at_1 = 0
hit_at_10 = 0

for i in results_dict:   
    id_query = i['query_idx']
    top_k_og = i['Top10_id']

    dist = i['dist']
    
    rank_list_dist_filter = [1 if x <= 3 else 0 for x in dist]
    if 1 not in rank_list_dist_filter:
        continue
    else:
        hits_clos = np.where(np.array(rank_list_dist_filter)[:10] == 1)[0]
        if hits_clos.size > 0:
            if hits_clos.size > 0:
                hit_at_10 += 1
                if hits_clos[0] == 0:
                    hit_at_1 += 1
                    
print("hit_at_1 ", hit_at_1, hit_at_1/len(results_dict))
print("hit_at_10 ", hit_at_10, hit_at_10/len(results_dict))
```
